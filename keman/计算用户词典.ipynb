```python
# coding=utf-8
import jieba
import re

from collections import Counter


class TextCutter(object):
    def __init__(self, stopwords):
        self.stop_word_list = stopwords
        self.zbj_word_list = []



    @staticmethod
    def get_word_list(filename):
        with open(filename, 'rb') as f:
            content = f.read().decode()
            words = content.splitlines()
            word_list = [w.strip() for w in words]
            return word_list

    def cut_text(self, text):

        words = jieba.cut(text, cut_all=False)
        counter = Counter()
        zbj_counter = Counter()
        for w in words:
            if w in self.stop_word_list:
                continue
            if w.isdigit():
                continue
            if len(w)<2:
                continue
            if w in self.zbj_word_list:
                zbj_counter[w] += 1
            counter[w] += 1

        print(counter.most_common())
        print(zbj_counter.most_common())

    def parse_phone(self):
        r = r"[0-9]+"
        phones = set(re.findall(r, self.text))
        print(phones)
```
