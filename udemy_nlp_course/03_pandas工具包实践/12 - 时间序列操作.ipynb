```python
import pymysql

pymysql.install_as_MySQLdb()
import pandas as pd 
from sqlalchemy import create_engine
sqlalchemy_uri = "mysql://root:mysql@localhost:3306/loopinsight"
engine = create_engine(sqlalchemy_uri)
sql = "select * from slices"
df = pd.read_sql_query(sql, engine, index_col='created_on')
```

```python
# 当指定created_on为时间索引之后， 就可以
# 1. 用pd.timestamp以切片的方式获取数据
df[pd.Timestamp('2018-01-02 08:55:00'):pd.Timestamp('2018-03-01 01:00:00')]
# 2. 也可以直接用数据，当create_on本身就是时间的话,可以直接时间
# 2.1 获取 2018 - 2019
df['2018':'2019']
# 2.2 获取 2018
df['2018']
# 2.3 获取以下时间切片的时间段
df['2018-01-01': '2018-03-02']
df['2017']
# 2.4. 获取以下时间切片月份段的
df['2018-02': '2018-03']
# 2.5 选取所有的一月份
df[df.index.month==1]
# 2.6 获取早上八点到十点的数据
df[(df.index.hour >=8) & (df.index.hour<=10)]
# 2.7 between time 
df.between_time('08:00', '12:00')
```

```python
# 计算以天，月，或者年为单位的平均值
df.resample('Y').mean().round(2).fillna(0)
df.resample('D').mean().round(2).fillna(0)
df.resample('M').mean().round(2).fillna(0)
```

```python
# 按照3天为单位进行统计
df.resample('3D').mean().round(2).fillna(0)
# 画图出来
import matplotlib
df.resample('M').mean().plot()
```
