```python
import pymysql
pymysql.install_as_MySQLdb()
import pandas as pd 
from sqlalchemy import create_engine
sqlalchemy_uri = "mysql://root:mysql@localhost:3306/dataview"
engine = create_engine(sqlalchemy_uri)
sql = "select * from res_manager_guanbao_business"
df = pd.read_sql_query(sql, engine)
```

```python
# 排序操作
df.sort_values(by=['income'], ascending=False)
# 对多个指标排序
# inplace=True的意思是将在原有的df上应用目前的操作，比如排序，比如groupby
df.sort_values(by=['systime', 'id'], ascending=[False, True], inplace=True)
```

```python
# 排序后去除掉重复的值， 但是去除重复的值的
data = df[['type_name', 'product_name']]

data.drop_duplicates().head()
```

```python
# 能否只要type_name里面有重复的值，就去除掉
data.drop_duplicates(subset='product_name')
```

```python
data.drop_duplicates(subset='type_name')
```

```python
# 使用map lambda函数，实现修改里面的元素
# map() 
def convert_type_name(name):
    return name.split('-')[-1]
data.type_name.map(lambda x:convert_type_name(x))

# 使用Map将营销转化为yingxiao
product_name_map = {
    "科技": 'tech'
}
# map操作的
data['product_name'].map(product_name_map)
```

```python
df.head()
```

```python
# apply函数 使用apply, 函数
import numpy as np 
def food_map(s):
    """传入一个series参数"""
    if s['product_name'] == '其他':
        return 'other'
    else:
        return s['product_name']

df.apply(func=food_map, axis='columns').head()
```

```python
# 使用Numpy随机构造一些数据
import numpy as np 
import pandas as pd 
data = {
    'data1': np.random.randn(5),
    'data2': np.random.randn(5),
}
df = pd.DataFrame(data)
```

```python
df
```

```python
# 添加一个列，这个列是其他两个列的计算
df = df.assign(ratio=df['data1']/df['data2'], tag=True)
```

```python
# 删除一个列
df
```

```python
df.drop('ratio', axis='columns', inplace=True)
```

```python
df
```

```python
# 创建一个series
s = pd.Series(np.random.randint(0, 100, 20))
s
```

```python
# 替换s中的东西
s.replace(97, np.nan, inplace=True)
```

```python
s.sort_values(inplace=True)
```

```python
s
```

```python
# 离散化处理
ages = np.random.randint(0, 50, 20)
bins = [10, 40, 80]

bins_res = pd.cut(ages, bins)
bins_res
```

```python
pd.value_counts(bins_res)
```

```python
pd.cut(ages, [10, 20, 30, 50, 80])
```

```python
groupNames = ['Young', 'Middle', 'Old']
s = pd.cut(ages, [10,30, 50, 80], labels=groupNames)

pd.value_counts(s)
```
