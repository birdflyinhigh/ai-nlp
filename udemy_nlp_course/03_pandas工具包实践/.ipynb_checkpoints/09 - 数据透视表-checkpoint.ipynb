```python
import pymysql

pymysql.install_as_MySQLdb()
```

```python
import pandas as pd 
from sqlalchemy import create_engine
sqlalchemy_uri = "mysql://root:mysql@localhost:3306/dataview"
engine = create_engine(sqlalchemy_uri)
```

```python
sql = "select * from res_manager_guanbao_business"
df = pd.read_sql_query(sql, engine)
```

```python
df.columns
```

```python
# 注意，默认是求平均，但是个人可以自定义其他的
pf = df.pivot_table(index='type_name', columns='area_name', values='income').round(2).fillna(0)
pf.head()
```

```python
pf.sum(axis="columns")
pf.sum(axis="index")
```

```python
df.columns
```

```python
# aggfunc是可以自定义的。常见的有max, min, count, sum, avg, mean, 默认是mean ,求平均值
df.pivot_table(index='product_name', columns='area_name', values='income', aggfunc='count').round(0).fillna(0)
```

```python
#与这个方法类似

pd.crosstab(index=df['product_name'], columns=df['area_name'])
```

```python
df.pivot_table(index='product_name', columns='province_name', values='income', aggfunc='count').round(0).fillna(0)
```

```python
df['Huiyuan'] = df['type_name'] =="会员费"
df.pivot_table(index='Huiyuan', columns='province_name', values='income', aggfunc='mean').round(2).fillna(0)
```
