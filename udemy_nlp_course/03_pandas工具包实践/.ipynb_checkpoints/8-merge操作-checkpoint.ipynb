````python
import pandas as pd 
````

````python
left_data = {
    'key': ['KO', 'K1', 'K2', 'K3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']
}

right_data = {
    'key': ['KO', 'K1', 'K2', 'K3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
}
````

````python
left = pd.DataFrame(data=left_data)
right = pd.DataFrame(data=right_data)
left
right
````

````python
pd.merge(left,right,on='key')
````

````python
left_data1 = {
    'key': ['KO', 'K1', 'K2', 'K3'],
    'key1': ['K1O', 'K11', 'K12', 'K13'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']
}

right_data1 = {
    'key': ['KO', 'K1', 'K2', 'K3'],
    'key1': ['K1O', 'K11', 'K12', 'K13'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
}
````

````python
left = pd.DataFrame(data=left_data1)
right = pd.DataFrame(data=right_data1)
pd.merge(left,right,on='key')
````

#### 假如指定多个Key, 假如key中的内容不一样，不一样的就会被删除

````python
left_data2 = {
    'key': ['KO', 'K1', 'K2', 'K3'],
    'key1': ['K1O', 'K11', 'K12', 'K4'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']
}

right_data2 = {
    'key': ['KO', 'K1', 'K2', 'K3'],
    'key1': ['K1O', 'K11', 'Kk', 'K13'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
}
left = pd.DataFrame(data=left_data2)
right = pd.DataFrame(data=right_data2)
# 不指定outer, 则为交集， 不同的删除掉
pd.merge(left,right,on=['key', 'key1'])
````

````python
pd.merge(left, right, how='outer', on=['key', 'key1'])
````
